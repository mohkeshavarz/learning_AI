```python
import numpy as np

```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4]).reshape(-1,1)
y=np.array([1,3,5,7])
```

```python
plt.plot(x,y,'*')
plt.show
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x,y)
```

```python
m = model.coef_[0]
c= model.intercept_

print(m,c)
```

```python
plt.plot(x,y,'*')
plt.plot(x,model.predict(x))
plt.show()
```

```python
result=model.predict([[5]])
result
```
